## Importação das bibliotecas

```python
# Tokenização
from nltk.tokenize import word_tokenize

# Stopwords
from nltk.corpus import stopwords

# POS Tagging
from nltk import pos_tag

# Stemming
from nltk.stem import PorterStemmer

# Lemmatization
from nltk.stem import WordNetLemmatizer

# Análise de sentimento
from nltk.sentiment import SentimentIntensityAnalyzer
```

## Tokenização
A **tokenização** é o processo de dividir um texto em partes menores, chamadas **tokens**. Esses tokens normalmente são palavras, mas também podem ser frases ou até caracteres. Essa é uma das primeiras etapas do Processamento de Linguagem Natural, pois permite que o computador analise o texto de forma organizada.

Exemplo prático.


```python
texto="O Processamento de Linguagem Natural é muito interessante!"
tokens=word_tokenize(texto)
print(tokens)
```

## Stopwords

As **stopwords** são palavras muito comuns em um idioma, como **"o"**, **"a"**, **"de"**, **"para"** e **"com"**. Em muitas aplicações de PLN, essas palavras são removidas porque geralmente não carregam informações importantes sobre o significado do texto.

Exemplo prático.

```python
texto="O processamento de linguagem natural é muito interessante."
tokens=word_tokenize(texto.lower())
sw=set(stopwords.words('portuguese'))

filtradas = []

for t in tokens:
  if t.isalpha() and t not in sw:
    filtradas.append(t)

print(filtradas)
```

## POS Tagging

O **POS Tagging (Part-of-Speech Tagging)** é o processo de identificar a **classe gramatical** de cada palavra em uma frase, como substantivo, verbo, adjetivo ou preposição. Isso ajuda o computador a compreender melhor a estrutura e o contexto do texto.

Exemplo prático.

```python
texto="The dog is running in the park."
tokens=word_tokenize(texto)
print(pos_tag(tokens))
```

## Vamos analisar palavra por palavra

| Palavra | Tag | Significado |
|----------|-----|-------------|
| **The** | **DT** | Determinante (artigo "o", "a", "the") |
| **dog** | **NN** | Substantivo |
| **is** | **VBZ** | Verbo **"to be"** (3ª pessoa do singular) |
| **running** | **VBG** | Verbo no gerúndio (terminação **-ing**) |
| **in** | **IN** | Preposição |
| **the** | **DT** | Determinante (artigo) |
| **park** | **NN** | Substantivo |
| **.** | **.** | Pontuação |


## Stemming

O **Stemming** reduz uma palavra ao seu **radical**, removendo prefixos ou sufixos. O objetivo é reduzir a palavra mesmo que o resultado nem sempre seja uma palavra existente no idioma.

Exemplo prático.

```python
stemmer=PorterStemmer()
palavras=["running","studies","connected","playing"]
for p in palavras:
    print(p,"->",stemmer.stem(p))
```

## Lemmatization

A **Lemmatization** também reduz palavras à sua forma básica, chamada **lema**, mas leva em consideração o contexto e a gramática da palavra. Por isso, costuma produzir resultados mais corretos do que o stemming.

Exemplo prático.

```python
lemma=WordNetLemmatizer()
palavras=["running","studies","mice","better"]
for p in palavras:
    print(p,"->",lemma.lemmatize(p))
print("Como verbo:")
print(lemma.lemmatize("running",pos="v"))
```

## Análise de Sentimento

A **Análise de Sentimento** é uma técnica que identifica a emoção ou opinião presente em um texto. Ela pode classificar uma frase como **positiva**, **negativa** ou **neutra**, sendo muito utilizada para analisar avaliações, comentários em redes sociais e opiniões de clientes.

Exemplo prático.

```python
sia=SentimentIntensityAnalyzer()
frases=[
"I love studying Artificial Intelligence!",
"This class is terrible.",
"The notebook is on the table."
]
for f in frases:
    print(f)
    print(sia.polarity_scores(f))
    print()
```
