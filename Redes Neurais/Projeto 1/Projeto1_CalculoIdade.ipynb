```python
# =====================================================================
# PASSO 1: Importar as bibliotecas
# =====================================================================
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

# =====================================================================
# PASSO 2: Criar os dados de treino
# =====================================================================
np.random.seed(42)
anos_nascimento = np.random.randint(1940, 2026, size=(150, 1)).astype(np.float32)
idades_reais = (2026 - anos_nascimento).astype(np.float32)

# --- O PULO DO GATO: NORMALIZAÇÃO ---
# Guardamos a média e o desvio padrão dos anos para usar no treino e no teste
ano_medio = anos_nascimento.mean()
ano_std = anos_nascimento.std()

# Transformando os anos em números pequenos (Média 0, Desvio Padrão 1)
anos_normalizados = (anos_nascimento - ano_medio) / ano_std

X_train = torch.from_numpy(anos_normalizados)
y_train = torch.from_numpy(idades_reais)

# =====================================================================
# PASSO 3: Estrutura do Perceptron
# =====================================================================
perceptron = nn.Linear(1, 1)
criterion = nn.MSELoss()
# Com dados normalizados, o otimizador funciona muito mais rápido e converge
optimizer = optim.Adam(perceptron.parameters(), lr=0.1)

# =====================================================================
# PASSO 4: Treinamento Eficiente
# =====================================================================
epochs = 2000  # Menos épocas são necessárias agora!

print("Treinando o neurônio atualizado...")
for epoch in range(epochs):
    optimizer.zero_grad()
    previsoes = perceptron(X_train)
    loss = criterion(previsoes, y_train)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 500 == 0:
        print(f"Época [{epoch+1}/{epochs}] -> Erro: {loss.item():.6f}")

print("\nTreinamento concluído com sucesso!")

# =====================================================================
# PASSO 5: Teste Interativo (Agora vai funcionar!)
# =====================================================================
print("\n--- PREVISÃO DO PERCEPTRON ---")
try:
    ano_teste = float(input("Digite o seu ano de nascimento: "))

    # ATENÇÃO: O ano digitado DEVE sofrer a mesma normalização que o treino sofreu
    ano_teste_normalizado = (ano_teste - ano_medio) / ano_std

    entrada = torch.tensor([[ano_teste_normalizado]], dtype=torch.float32)

    perceptron.eval()
    with torch.no_grad():
        idade_prevista = perceptron(entrada).item()

    print("-" * 45)
    print(f"Previsão do Perceptron: {round(idade_prevista)} anos.")
    print(f"Cálculo matemático exato (2026 - {int(ano_teste)}): {2026 - int(ano_teste)} ano(s).")
    print("-" * 45)

except ValueError:
    print("Por favor, digite um ano válido.")
```
